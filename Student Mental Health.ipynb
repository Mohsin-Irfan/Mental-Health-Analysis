#### A STATISTICAL RESEARCH ON THE EFFECTS OF MENTAL HEALTH ON STUDENTSâ€™ CGPA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student Mental health.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.nunique()
```

```python
df = df.dropna(how='any',axis=0)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,10))
plt.hist(df['Age'],color=['gold'])
plt.title('Age Distribution')
plt.show()
```

```python
df.rename(columns={'Choose your gender': 'Gender'},inplace=True)
```

```python
df.head()
```

```python
df.Gender.value_counts()
```

```python
plt.figure(figsize=(12,6))
plt.title("Gender Distribution")
plt.pie(df.Gender.value_counts(),  explode=(0.025,0.025), labels=df.Gender.value_counts().index, 
            colors=['skyblue','lightpink'], autopct='%1.1f%%', startangle=180)
plt.legend()
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(10, 10))
sns.countplot(x='Your current year of Study', hue='Gender', data=df)
plt.title("Students studying in a particular year")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x = 'Do you have Anxiety?', hue='Do you have Depression?', data = df)
plt.title("Students studying in a particular year")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.set_theme(style='darkgrid')
sns.countplot(y= "Do you have Anxiety?", hue="Gender", data=df)
plt.title("Anxiety by Gender")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.set_theme(style='whitegrid')
sns.countplot(y="Do you have Depression?", hue="Gender", data=df)
plt.title("Depression by Gender")
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(10,10))
sns.set_theme(style='darkgrid')
sns.countplot(x= "Do you have Anxiety?", hue="Your current year of Study", data=df)
plt.title("Anxiety by study year")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.set_theme(style='darkgrid')
sns.countplot(x= "Do you have Depression?", hue="Your current year of Study", data=df)
plt.title("Depression by study year")
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(10,10))
sns.set_theme(style='darkgrid')
sns.countplot(x= "Do you have Panic attack?", hue="What is your CGPA?", data=df)
plt.title("Panic attack by CGPA")
plt.show()
```
